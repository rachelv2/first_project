```python
#Best way giving aging_score
#This gives you a weighted aging score where 80+ counts 4x more than 65-69, so countries with truly old populations bubble to the top!
import polars as pl

merged = pl.read_csv('merged.csv')

# Step 1: Indicators that already have Total
has_total = merged.filter(
    (pl.col('INDICATOR_LABEL').is_in([
        'Population ages 65 and above (% of total population)',
        'Population ages 65 and above, total'
    ])) &
    (pl.col('SEX_LABEL') == 'Total')
)

# Step 2: Average Male+Female for granular bands
no_total = (
    merged
    .filter(
        pl.col('INDICATOR_LABEL').is_in([
            'Population ages 80 and above, male (% of male population)',
            'Population ages 80 and above, female (% of female population)',
            'Population ages 75-79, male (% of male population)',
            'Population ages 75-79, female (% of female population)',
            'Population ages 70-74, male (% of male population)',
            'Population ages 70-74, female (% of female population)',
            'Population ages 65-69, male (% of male population)',
            'Population ages 65-69, female (% of female population)',
        ])
    )
    .group_by(['REF_AREA', 'REF_AREA_LABEL', 'AGE_LABEL'])
    .agg([
        pl.col('2018_hdi').mean().alias('2018_hdi'),
        pl.col('2010_hdi').mean().alias('2010_hdi'),
        pl.lit('Total').alias('SEX_LABEL'),
        pl.col('INDICATOR_LABEL').first()
    ])
)

# Step 3: Build aging score — higher weight for older bands
aging_score = (
    pl.concat([has_total, no_total], how='diagonal')
    .group_by(['REF_AREA', 'REF_AREA_LABEL'])
    .agg([
        (
            pl.col('2018_hdi').filter(pl.col('AGE_LABEL') == '80 years old and over').sum() * 4 +
            pl.col('2018_hdi').filter(pl.col('AGE_LABEL') == '75 to 79 years old').sum() * 3 +
            pl.col('2018_hdi').filter(pl.col('AGE_LABEL') == '70 to 74 years old').sum() * 2 +
            pl.col('2018_hdi').filter(pl.col('AGE_LABEL') == '65 to 69 years old').sum() * 1
        ).alias('aging_score')
    ])
    .sort('aging_score', descending=True)
)

# Top 3 oldest and bottom 3 youngest
top3 = aging_score.head(3)
bottom3 = aging_score.tail(3)
top_bottom = pl.concat([top3, bottom3])

display(top_bottom)
```
