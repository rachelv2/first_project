Ranking countries from oldest to youngest populations (and vice-versa) in 2010 and 2018

```python
import polars as pl

wdi = pl.read_csv("../data/raw/WB_WDI_WIDEF.csv")

wdi.head()
```

```python
indicators = (
    wdi
    .select("INDICATOR_LABEL")
    .unique()
    .sort("INDICATOR_LABEL")
)

for i in indicators["INDICATOR_LABEL"]:
    print(i)
```

```python
# Keep only total-sex rows and cast years

wdi = wdi.filter(pl.col("SEX_LABEL") == "Total").with_columns([
    pl.col("2010").cast(pl.Float64),
    pl.col("2018").cast(pl.Float64),
])
```

```python
# Total population

total_pop = (
    wdi
    .filter(pl.col("INDICATOR_LABEL") == "Population, total")
    .select([
        pl.col("REF_AREA_LABEL").alias("country"),
        pl.col("2010").alias("total_population_2010"),
        pl.col("2018").alias("total_population_2018"),
    ])
    .unique(subset=["country"])
)
```

```python
# Population 65+

pop65 = (
    wdi
    .filter(pl.col("INDICATOR_LABEL") == "Population ages 65 and above, total")
    .select([
        pl.col("REF_AREA_LABEL").alias("country"),
        pl.col("2010").alias("population_65plus_2010"),
        pl.col("2018").alias("population_65plus_2018"),
    ])
    .unique(subset=["country"])
)
```

```python
# Build final ranking table in one pipeline, oldest country to youngest country

ranking_clean = (
    total_pop
    .join(pop65, on="country", how="inner")
    .with_columns([
        ((pl.col("population_65plus_2010") / pl.col("total_population_2010")) * 100).round(3).alias("share_65plus_2010"),
        ((pl.col("population_65plus_2018") / pl.col("total_population_2018")) * 100).round(3).alias("share_65plus_2018"),
        (pl.col("total_population_2010") / 1_000_000).round(3).alias("total_pop_2010_millions"),
        (pl.col("population_65plus_2010") / 1_000_000).round(3).alias("pop_65plus_2010_millions"),
        (pl.col("total_population_2018") / 1_000_000).round(3).alias("total_pop_2018_millions"),
        (pl.col("population_65plus_2018") / 1_000_000).round(3).alias("pop_65plus_2018_millions"),
    ])
    .sort("share_65plus_2018", descending=True)
    .with_row_index("rank_2018", offset=1)
    .select([
        "rank_2018",
        "country",
        "total_pop_2010_millions",
        "pop_65plus_2010_millions",
        "share_65plus_2010",
        "total_pop_2018_millions",
        "pop_65plus_2018_millions",
        "share_65plus_2018",
    ])
)

ranking_clean.head(20) # Oldest countries at the top!
```

```python
# Youngest countries at the bottom!

ranking_clean.tail(20) 
```

NOTES: consider excluding micro states like Monaco, San Marino, Liechtenstein and Andorra

NEW INDICATOR: "share_65plus" represents the proportion of a country’s population that is aged 65 years or older

Share 65+	Interpretation
25–30%	Very old population (Japan)
18–23%	Aging developed countries (Italy, Germany)
10–15%	Transitional economies
2–5%	Very young populations (many African countries)

```python
# NEW INDICATOR: adding “aging speed” to reveal which countries are aging the fastest, not just which are already old

ranking_clean = ranking_clean.with_columns(
    (pl.col("share_65plus_2018") - pl.col("share_65plus_2010"))
    .round(2)
    .alias("aging_speed_2010_2018")
)
```

```python
# FINAL 2ND VISUALIZATION

ranking_no_micro = ranking_clean.filter(~pl.col("country").is_in(excluded))

# 5 oldest and 5 youngest
oldest_5 = ranking_no_micro.sort("share_65plus_2018", descending=True).head(5)
youngest_5 = ranking_no_micro.sort("share_65plus_2018").head(5)

comparison = pl.concat([oldest_5, youngest_5]).sort("share_65plus_2018")

countries = comparison["country"].to_list()
values_2010 = comparison["share_65plus_2010"].to_list()
values_2018 = comparison["share_65plus_2018"].to_list()

old_countries = oldest_5["country"].to_list()

y = np.arange(len(countries))

fig, ax = plt.subplots(figsize=(11.5,7))
fig.patch.set_facecolor("white")
ax.set_facecolor("white")

for i, country in enumerate(countries):

    is_old = country in old_countries

    # Colors
    if is_old:
        line_color = "#1f77b4"     # blue
        dot_2018_color = "#1f77b4"
    else:
        line_color = "#f2a65a"     # warm orange
        dot_2018_color = "#f2a65a"

    dot_2010_color = "#888888"

    line_width = 3.2 if is_old else 2.6
    dot_size = 65 if is_old else 55

    # connecting line
    ax.hlines(
        y=i,
        xmin=values_2010[i],
        xmax=values_2018[i],
        color=line_color,
        linewidth=line_width,
        zorder=1
    )

    # dots
    ax.scatter(values_2010[i], i, color=dot_2010_color, s=dot_size, zorder=2)
    ax.scatter(values_2018[i], i, color=dot_2018_color, s=dot_size, zorder=3)

    # value labels (more spacing)
    ax.text(
        values_2010[i] - 0.7,
        i,
        f"{values_2010[i]:.1f}%",
        va="center",
        ha="right",
        fontsize=10,
        color="#666666"
    )

    ax.text(
        values_2018[i] + 0.7,
        i,
        f"{values_2018[i]:.1f}%",
        va="center",
        ha="left",
        fontsize=10.5,
        color="#222222"
    )

# country labels
ax.set_yticks(y)
ax.set_yticklabels(countries, fontsize=11)

# breathing room on both sides
xmin = min(values_2010) - 3.2
xmax = max(values_2018) + 3.2
ax.set_xlim(xmin, xmax)

# titles
ax.set_title(
    "The World's Demographic Divide, 2010–2018",
    fontsize=19,
    loc="left",
    pad=50
)

ax.text(
    0,1.04,
    "Older societies continued aging fast, while the youngest populations remained extremely young.",
    transform=ax.transAxes,
    fontsize=11,
    color="#555555"
)

ax.set_xlabel("Population aged 65+ (%)", fontsize=12, labelpad=20, loc="left")

# styling
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["left"].set_visible(False)
ax.spines["bottom"].set_color("#cccccc")

ax.grid(axis="x", linestyle="--", linewidth=0.8, alpha=0.18)

ax.tick_params(axis="y", length=0)
ax.tick_params(axis="x", colors="#666666")


plt.tight_layout()
plt.show()
```

```python
# FINAL 3RD VISUALIZATION

quadrant = (
    ranking_clean
    .with_columns(
        (pl.col("share_65plus_2018") - pl.col("share_65plus_2010"))
        .round(2)
        .alias("aging_speed")
    )
    .filter(~pl.col("country").is_in(excluded))
)

countries = quadrant["country"].to_list()
aging_level = quadrant["share_65plus_2018"].to_list()
aging_speed = quadrant["aging_speed"].to_list()

# Identify oldest and youngest groups
oldest = quadrant.sort("share_65plus_2018", descending=True).head(6)["country"].to_list()
youngest = quadrant.sort("share_65plus_2018").head(3)["country"].to_list()
middle = quadrant.sort("share_65plus_2018").slice(len(quadrant)//2 - 4, 2)["country"].to_list()

# Median lines
x_mid = quadrant["share_65plus_2018"].median()
y_mid = quadrant["aging_speed"].median()

fig, ax = plt.subplots(figsize=(12,7))
fig.patch.set_facecolor("white")
ax.set_facecolor("white")

for i, country in enumerate(countries):

    if country in oldest:
        color = "#1f77b4"  # blue
        size = 70
    elif country in youngest:
        color = "#f2a65a"  # warm orange
        size = 70
    elif country in middle:
        color = "#9d69c7"
        size = 60
    else:
        color = "#bfbfbf"
        size = 30
    

    ax.scatter(
        aging_level[i],
        aging_speed[i],
        s=size,
        color=color,
        alpha=0.9,
        zorder=3
    )

    if country in oldest or country in youngest:
        ax.text(
            aging_level[i] + 0.2,
            aging_speed[i] + 0.02,
            country,
            fontsize=10,
            color="#333333"
        )

    if country in middle:
        ax.text(
            aging_level[i] + 0.2,
            aging_speed[i] - 0.04,
            country,
            fontsize=10,
            color="#555555"
        )

# Quadrant reference lines
ax.axvline(x=x_mid, color="#999999", linestyle="--", linewidth=1)
ax.axhline(y=y_mid, color="#999999", linestyle="--", linewidth=1)

# Labels
ax.set_xlabel("Population aged 65+ in 2018 (%)", fontsize=12, labelpad=20, loc="left")
ax.set_ylabel("Increase in population aged 65+ (2010–2018)", fontsize=12)

ax.set_title(
    "Aging Level vs Aging Speed",
    fontsize=20,
    loc="left",
    pad=50
)

ax.text(
    0,1.04,
    "Some countries are already old, while others are aging rapidly and may face future labor-market pressure.",
    transform=ax.transAxes,
    fontsize=11,
    color="#555555"
)

# Styling
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["left"].set_color("#cccccc")
ax.spines["bottom"].set_color("#cccccc")

ax.grid(True, linestyle="--", linewidth=0.8, alpha=0.15)

ax.tick_params(colors="#666666")


plt.tight_layout()
plt.show()
```
