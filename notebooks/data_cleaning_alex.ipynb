```python
import pandas as pd
import numpy as np
import yaml

with open("../config.yaml", "r") as f:
    config = yaml.safe_load(f)

#config["input_data"]["file_alex"]
hci_df = pd.read_csv(config["input_data"]["file_alex"])
```

```python
# we change the name of the columns in lower case.
hci_df.columns = hci_df.columns.str.lower()
hci_df.head()
```

```python
# we will filter the dataframe to only keep the indicators: "WB_HCI_HCI", "WB_HCI_LAYS" and "WB_HCI_ASR"
hci_df = hci_df[hci_df["indicator"].isin(["WB_HCI_HCI", "WB_HCI_LAYS", "WB_HCI_ASR"])]
hci_df
```

```python
# sort the dataframe by "ref_area" and "time_period" in ascending order
hci_df= hci_df.sort_values(by=["ref_area", "time_period"] , ascending=True)
hci_df
```
